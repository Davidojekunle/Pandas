```python
data_set['SEX'] =data_set['SEX'].astype(str)
data_set['SEX'], data_set['SEX']  = data_set['SEX'].str.replace('1', 'Female', regex=True),data_set['SEX'].str.replace('2', 'Male', regex=True)
summer = data_set.head(30)
summer.plot.bar(x='SEX', y='AGE')
```

```python
data = data_set.head(1000)
data.groupby('DATE_DIED')['AGE'].plot(kind='hist', alpha=0.5, legend=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Death')
plt.show()
```

```python
data_set.groupby('PREGNANT')['AGE'].plot(kind='hist', alpha=0.5, legend=True)
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.title('Age Distribution by Pregnancy Status')
plt.show()
```

```python
dead = data_set[data_set["DATE_DIED"]!= '9999-99-99']
not_dead =  data_set[data_set['DATE_DIED']=='9999-99-99']

plt.figure(figsize=(10, 6))
dead[dead['SEX']=='Male']['AGE'].plot(kind='hist', alpha=0.5, label='Male (Died)', color='blue', legend=True, bins= 30)
dead[dead['SEX']=='Female']['AGE'].plot(kind='hist',alpha=0.5, label='Female (Died)', color='red' ,legend=True, bins= 30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Patients who Died by Sex')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
not_dead[not_dead['SEX']=='Male']['AGE'].plot(kind='hist', alpha=0.5, label='Male (Did not Die)', legend=True, bins= 30, edgecolor= 'black')
not_dead[not_dead['SEX']=='Female']['AGE'].plot(kind='hist',alpha=0.5, label='Female (Did not Died)',legend=True, bins= 30,  edgecolor= 'blue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Patients who  did not die by Sex')
plt.legend()
plt.show()
```

```python
data_set.groupby('PATIENT_TYPE')['AGE'].plot(kind='hist',alpha=0.5, legend= True, edgecolor='black')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Patients who Were Discharged')
plt.show()
```

```python

labels = ['Discharged', 'Not Discharged']
counts = data_set['PATIENT_TYPE'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=140)
plt.title('Percentage of Patients Discharged vs Not Discharged')
plt.axis('equal') 

plt.show()
```

```python
dead = data_set[data_set["DATE_DIED"]!= '9999-99-99']
not_dead =  data_set[data_set['DATE_DIED']=='9999-99-99']

counts = [len(dead), len(not_dead)]
plt.figure(figsize=(8, 8))
plt.pie(counts,  autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=140)
plt.title('dead against not dead')
plt.axis('equal') 
plt.show()
```
