```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Covid_data[1].csv')
data['SEX'] = data['SEX'].astype(str)
data['PATIENT_TYPE'] = data['PATIENT_TYPE'].astype(str)
```

```python
data['SEX'] =  data['SEX'].str.replace('2', 'Male', regex= True)
data['PATIENT_TYPE'] =  data['PATIENT_TYPE'].str.replace('2', 'Hospitalized', regex= True)
data['PATIENT_TYPE'] = data['PATIENT_TYPE'].str.replace('1', 'Returned Home', regex= True)
data['SEX'] = data['SEX'].str.replace('1', 'Female', regex=True)
dataset = data.head(100000)
```

```python
died = dataset[dataset['DATE_DIED'] != '9999-99-99']
not_dead = dataset[dataset['DATE_DIED'] == '9999-99-99']

not_dead[not_dead['SEX'] =='Male']['AGE'].plot(kind='hist', alpha=0.5, label='Male (not_Dead)', color='blue', legend=True, edgecolor='black')
not_dead[not_dead['SEX'] =='Female']['AGE'].plot(kind='hist', alpha = 0.5, label= 'Female(not_dead)',legend=True, edgecolor= 'black', color= 'orange')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.title('Age distrubution of Dead Patient(Male/Female)')
plt.show()
```

```python
not_dead[not_dead['SEX'] =='Male']['AGE'].plot(kind='hist', alpha = 0.5, edgecolor= 'black', label= 'Male not dead', legend=True)
died[died['SEX'] =='Male']['AGE'].plot(kind='hist', alpha = 0.5, edgecolor = 'black', label= 'Male dead', legend=True)
plt.xlabel('Male Age')
plt.ylabel('Frequency')
plt.legend()
plt.title('MAle Age distribution of non dead/dead patients')
plt.show()
```

```python
not_dead[not_dead['SEX'] =='Female']['AGE'].plot(kind='hist', alpha = 0.5, edgecolor= 'black', label= 'Female not dead', legend=True)
died[died['SEX'] =='Female']['AGE'].plot(kind='hist', alpha = 0.5, edgecolor = 'black', label= 'Female dead', legend=True)
plt.xlabel('Female Age')
plt.ylabel('Frequency')
plt.legend()
plt.title('Female Age distribution of non dead/dead patients')
plt.show()
```

```python
dataset.groupby('PATIENT_TYPE')['AGE'].plot(kind='hist',alpha=0.5, legend= True, edgecolor='black')
plt.xlabel('Age')
plt.title('Age distribution of patients types')
```

```python
data.groupby('DIABETES')['AGE'].plot(kind='hist',alpha=0.3, legend= True, edgecolor='black')
plt.xlabel('Age')
plt.title('Age distribution of patients with Diebetes')
```

```python
died = data[data['DATE_DIED'] != '9999-99-99']
not_dead = data[data['DATE_DIED'] == '9999-99-99']

dead_females = died[died['SEX'] =='Female']
dead_males = died[died['SEX'] =='Male']
lables = ['dead_females', 'dead_males']
counts =[len(dead_females), len(dead_males)]
plt.pie(counts,  autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=140, labels=lables)
plt.title('PIe chart of dead females against dead males')
plt.show()
```

```python
not_dead = data[data['DATE_DIED'] == '9999-99-99']

not_dead_females = not_dead[not_dead['SEX'] =='Female']

not_dead_males = not_dead[not_dead['SEX'] =='Male']

lables = ['living_females', 'living_males']

counts =[len(not_dead_females), len(not_dead_males)]
plt.pie(counts,  autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=140, labels=lables)
plt.title('PIe chart of dead females against dead males')
plt.show()
```
